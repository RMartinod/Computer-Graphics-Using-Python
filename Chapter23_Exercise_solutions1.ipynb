**E.4.1.** The Olympic rings are five rings representing the inhabited continents of the world. It was made to contain the colors (blue, black, red, yellow and green) that are common to almost all flags of the world.

```python
import numpy as np, matplotlib.pyplot as plt
u = np.linspace(0,2*np.pi,100)
x = np.cos(u)
y = np.sin(u)
C1 = np.array([x,y,np.ones(len(x))])
Dp = [[0.0,2.2,4.4, 1.1, 3.3],
      [0.0,0.0,0.0,-1.0,-1.0],
      [1,1,1,1,1,1]]
colorRing = ['blue','black','red','yellow','green']
for ring in range(len(Dp[0])):
  TM = np.array([[1, 0, Dp[0][ring]],
                 [0, 1, Dp[1][ring]],
                 [0, 0, Dp[2][ring]]])
  C2 = TM@C1
  plt.plot(C2[0,:],C2[1,:],color=colorRing[ring],linewidth=6)
```

**E.4.3.** A honeycomb is a mass of hexagonal prismatic cells built from beeswax by honeybees in their nests to contain their brood (eggs, larvae, and pupae) and stores of honey and pollen.

There are two main explanations for why honeycombs are composed of hexagons rather than any other shape. The first, known as the honeycomb conjecture, suggests that hexagonal tiling optimizes the partitioning of space by creating equal-sized cells while minimizing the total perimeter of those cells. This arrangement is the most efficient way to divide a given area into cells with the least amount of material. This conjecture was proposed by Jan Brożek and was mathematically proven by Thomas Hales. In this way, the hexagonal structure minimizes the material required to construct a lattice of cells within a given volume. A second explanation, proposed by D'Arcy Wentworth Thompson, suggests that the hexagonal pattern arises naturally from the process by which individual bees construct the cells. This process is similar to the formation of the boundary shapes in a cluster of soap bubbles. Thompson noted that queen cells, which are built individually by bees, are often irregular and lumpy, lacking the geometric efficiency seen in the honeycomb, supporting the idea that the hexagonal shape emerges from the bees' natural building behavior.


```python
import numpy as np, matplotlib.pyplot as plt
u   = np.linspace(0,2*np.pi,7)
x   = np.cos(u)
y   = np.sin(u)
S1  = np.array([x,y,np.ones(len(x))])
Dp = [[0,1.50,3.0,4.50,0.00,1.50,3.00,4.50],
      [0,0.85,0.0,0.85,1.73,2.58,1.73,2.58],
      [1,1,1,1,1,1,1,1]]
for cell in range(len(Dp[0])):
  TM = np.array([[1, 0, Dp[0][cell]],
                 [0, 1, Dp[1][cell]],
                 [0, 0, Dp[2][cell]]])
  S2 = TM@S1
  plt.fill(S2[0,:],S2[1,:],color='orange',alpha=.8)
```

**E.4.4.** Polar coordinates or polar coordinate system is a two-dimensional coordinate system in which each point in the plane is determined by a distance and an angle. This system is widely used in physics and trigonometry.

```python
import numpy as np, matplotlib.pyplot as plt
u   = np.linspace(0,2*np.pi,80)
v   = np.linspace(1,4,5)
U,V = np.meshgrid(u,v)
r   = V + np.cos(8*U)
X   = r * np.sin(U)
Y   = r * np.cos(U)
m,n = np.shape(X)
S1 = np.ones((3,m*n))
for row in range(m):
  S1[0:2,row*n:(row+1)*n] = [X[row,:],Y[row,:]]
Dp = [[0,10,20,0,10,20,0,10,20],[0,0,0,10,10,10,20,20,20],[1,1,1,1,1,1,1,1,1]]
for tile in range(len(Dp[0])):
  TM = np.array([[1, 0, Dp[0][tile]],
                 [0, 1, Dp[1][tile]],
                 [0, 0, Dp[2][tile]]])
  S2 = TM@S1
  X2 = np.zeros((m,n))
  Y2 = np.zeros((m,n))
  for row in range(m):
    X2[row,:] = S2[0,row*n:(row+1)*n]
    Y2[row,:] = S2[1,row*n:(row+1)*n]
  for row in range(m-1):
    for col in range(n-1):
      plt.fill([X2[row,col],X2[row+1,col],X2[row+1,col+1],X2[row,col+1]],
               [Y2[row,col],Y2[row+1,col],Y2[row+1,col+1],Y2[row,col+1]],
               color='blue',alpha=.3)
```

**E.5.2.** We give you a code that models and plots a flower petal. Modify the code to model and plot a flower with twelve petals by applying rotation transformations

```python
import numpy as np, matplotlib.pyplot as plt
u = np.linspace(0,100*np.pi,10000)
x = 3*u/(1+u**3)
y = 3*u**2/(1+u**3)
S1 = np.array([x,y,np.ones(len(x))])
angle = 30*np.pi/180
ax = plt.subplots()
for petal in range(13):
   theta = petal*angle
   RM = np.array([[np.cos(theta), -np.sin(theta), 0],
                  [np.sin(theta),  np.cos(theta), 0],
                  [0, 0, 1]])
   S2 = RM @ S1
   plt.fill(S2[0,:],S2[1,:], color='salmon', alpha=.6)
```

**E.6.1.** Chanel is one of the brands that has stood the test of time with only one logo. Chanel's main emblem and original design is a monogram logo. It made its official appearance in the 1920s. At first glance, we see two “Cs”, one of which reflects the other and are embedded in each other. Below, we give you a code that models and plots a “C” element of the Chanel logo.

```python
import numpy as np, matplotlib.pyplot as plt
u   = np.linspace(3*np.pi/10,-13*np.pi/10,50)
v   = np.linspace(3.5,5,2)
U,V = np.meshgrid(u,v)
X   = V * np.sin(U) + 2.5
Y   = V * np.cos(U)
m,n = np.shape(X)
S1 = np.ones((3,m*n))
for row in range(m):
  S1[0:2,row*n:(row+1)*n] = [X[row,:],Y[row,:]]
RfM = np.array([[-1, 0, 0],
                [ 0, 1, 0],
                [ 0, 0, 1]])
S2 = RfM@S1
X2, Y2 = np.zeros((m,n)), np.zeros((m,n))
for row in range(m):
  X2[row,:] = S2[0,row*n:(row+1)*n]
  Y2[row,:] = S2[1,row*n:(row+1)*n]
for row in range(m-1):
  for col in range(n-1):
    plt.fill([X[row,col],X[row+1,col],X[row+1,col+1],X[row,col+1]],
             [Y[row,col],Y[row+1,col],Y[row+1,col+1],Y[row,col+1]],
              color='goldenrod',alpha=1)
    plt.fill([X2[row,col],X2[row+1,col],X2[row+1,col+1],X2[row,col+1]],
             [Y2[row,col],Y2[row+1,col],Y2[row+1,col+1],Y2[row,col+1]],
              color='goldenrod',alpha=1)
```

**E.6.2.** Do you identify the Renault logo? Renault adopts the diamond; this geometrical symbol was definitively adopted in 1925. In the first instance, the diamond was used exclusively on luxury sports vehicles identified by the name Stella from 1929. Below, we give you a code that models and plots a section element of the Renault logo.

```python
import numpy as np, matplotlib.pyplot as plt
x  = [-1.5,-0.5, 0.5, -0.5]
y  = [ 0, 0,  -2,   -2]
S1 = np.array([x,y,np.ones(len(x))])
RfM1 = np.array([[-1, 0, 0],
                 [ 0, 1, 0],
                 [ 0, 0, 1]])
RfM2 = np.array([[ 1, 0, 0],
                 [ 0,-1, 0],
                 [ 0, 0, 1]])
S2 = RfM1 @ S1
S3 = RfM2 @ S1
S4 = RfM1 @ S3
plt.fill(S2[0,:],S2[1,:],color='silver',alpha=0.95)
plt.fill(S3[0,:],S3[1,:],color='silver',alpha=0.95)
plt.fill(S1[0,:],S1[1,:],color='darkgrey',alpha=0.95)
plt.fill(S4[0,:],S4[1,:],color='gainsboro',alpha=0.95)
plt.axis('equal')
```

**E.7.1.** We give you a code that models and plots a heart. Modify the code to model and plot a sequence of hearts by applying a uniform scaling series, which shrinks the hearts on the x-axis and y-axis.

```python
import numpy as np, matplotlib.pyplot as plt
u = np.linspace(0,2*np.pi,1000)
x = 16*np.sin(u)**3
y = 5+13*np.cos(u)-5*np.cos(2*u)-2*np.cos(3*u)-np.cos(4*u)
S1 = np.array([x,y,np.ones(len(x))])
scaling = np.linspace(1,10,10)/10
for scale in scaling:
  SM = np.array([[scale,    0, 0],
                 [0,    scale, 0],
                 [0,        0, 1]])
  S2 = SM@S1
  plt.fill(S2[0,:],S2[1,:],color='darkred',alpha=1-scale)
plt.axis('equal')
```

**E.7.2.** We give you a code that models and plots an ellipsoid. Modify the code to model and plot a sequence of ellipsoids by applying a non-uniform scaling series, which shrinks the ellipsoid on the x-axis and stretches the ellipsoid on the y-axis

```python
import numpy as np, matplotlib.pyplot as plt
th = np.linspace(0,2*np.pi,50)
x = 5 * np.cos(th)
y = np.sin(th)
S1 = np.array([x,y,np.ones(len(x))])
scaling = np.linspace(1,5,10)
for scale in scaling:
  SM = np.array([[1/scale,    0, 0],
                 [0,      scale, 0],
                 [0,          0, 1]])
  S2 = SM@S1
  plt.fill(S2[0,:],S2[1,:],color='darkgreen',alpha=.3)
plt.axis('equal')
```

**E.8.1.** We give you a code that models and plots a 3D-hypocycloid model. Modify the code to model and plot a set of eight (8) stacked 3D-hypocycloids (side by side) by applying translation

```python
import numpy as np, plotly.graph_objects as go
u, v = np.linspace(0,2*np.pi,60), np.linspace(0,np.pi,60)
U, V = np.meshgrid(u,v)
X = 0.8*(np.cos(U)**3*np.sin(V)**3)+0.2*(np.cos(U)*np.sin(V))
Y = 0.8*(np.sin(U)**3*np.sin(V)**3)+0.2*(np.sin(U)*np.sin(V))
Z = 0.8*np.cos(V)**3+0.2*np.cos(V)
fig = go.Figure(go.Surface(x=X,y=Y,z=Z,opacity=1,colorscale ='amp'))
m,n = np.shape(X)
S1 = np.ones((4,m*n))
for row in range(m):
  S1[0:3,row*n:(row+1)*n] = [X[row,:],Y[row,:],Z[row,:]]
Dp = [[2,0,0,1],[0,2,0,1],[2,2,0,1],[0,0,2,1],[2,0,2,1],[0,2,2,1],[2,2,2,1]]
for i in range(len(Dp)):
  TM = np.array([[1,0,0,Dp[i][0]],[0,1,0,Dp[i][1]],[0,0,1,Dp[i][2]],[0,0,0,Dp[i][3]]])
  S2 = TM@S1
  X2, Y2, Z2 = np.zeros((m,n)), np.zeros((m,n)), np.zeros((m,n))
  for row in range(m):
    X2[row,:] = S2[0,row*n:(row+1)*n]
    Y2[row,:] = S2[1,row*n:(row+1)*n]
    Z2[row,:] = S2[2,row*n:(row+1)*n]
  fig.add_trace(go.Surface(x=X2,y=Y2,z=Z2,opacity=1,colorscale='amp'))
fig.update_layout(scene=dict(aspectratio=dict(x=1, y=1, z=1)))
```
