**Reflection for a local homogeneous coordinate system in 2D**

```python
import numpy as np, plotly.express as px
LC1 = np.array([[ 0.866,-0.500, 1.732],
                [ 0.500, 0.866, 1.000],
                [ 0.000, 0.000, 1.000]])
RfM = np.array([[-1, 0, 0],
                [ 0, 1, 0],
                [ 0, 0, 1]])
LC2 = np.matmul(RfM,LC1)
fig = px.line(x=[LC1[0,0]+LC1[0,2], LC1[0,2], LC1[0,1]+LC1[0,2]],
              y=[LC1[1,0]+LC1[1,2], LC1[1,2], LC1[1,1]+LC1[1,2]])
fig.add_trace(px.line(x=[LC2[0,0]+LC2[0,2], LC2[0,2], LC2[0,1]+LC2[0,2]],
                      y=[LC2[1,0]+LC2[1,2], LC2[1,2], LC2[1,1]+LC2[1,2]]).data[0])
fig.update_traces(line_color='red',line_width=4)
```

**Reflection for a curve in 2D**

```python
import numpy as np, plotly.express as px
u = np.linspace(0,2*np.pi,200)
x = (1 + 2*np.sin(3*u)) * np.cos(u)
y = (1 + 2*np.sin(3*u)) * np.sin(u) + 3.5
C1 = np.array([x,y,np.ones(len(x))])
RfM = np.array([[ 1, 0, 0],
                [ 0,-1, 0],
                [ 0, 0, 1]])
C2 = np.matmul(RfM,C1)
fig = px.line(x=C1[0,:],y=C1[1,:])
fig.add_trace(px.line(x=C2[0,:],y=C2[1,:]).data[0])
fig.update_traces(line_color='black',line_width=4)
fig.update_layout(title='Butterfly',autosize=False)
```

**Reflection for a surface in 2D**

```python
import numpy as np, matplotlib.pyplot as plt
u = np.linspace(0,6*np.pi,200)
x = (8-3) * np.cos(u) + 3 * np.cos((8-3)*u/3) + 6
y = (8-3) * np.sin(u) - 3 * np.sin((8-3)*u/3) + 6
S1 = np.array([x,y,np.ones(len(x))])
RfM = np.array([[-1, 0, 0],
                [ 0,-1, 0],
                [ 0, 0, 1]])
S2 = RfM @ S1
plt.fill(S1[0,:],S1[1,:],color='darkblue',alpha=.8)
plt.fill(S2[0,:],S2[1,:],color='darkgreen',alpha=.8)
```

```python
import numpy as np, matplotlib.pyplot as plt
th = np.linspace(0,2*np.pi,50)
x = np.cos(th)
y = np.sin(th)
S1 = np.array([x,y,np.ones(len(x))])
plt.fill(S1[0,:],S1[1,:],color='green',alpha=0.8)
u = np.linspace(-3*np.pi/4,3*np.pi/4,200)
x = 2*u
Ca = np.array([x, (1/(u**2 + 0.8))-0.1, np.ones(len(x))])
Cb = np.array([x, (1/(u**2 + 0.9))+0.2, np.ones(len(x))])
Cc = np.array([x, (1/(u**2 + 1.0))+0.5, np.ones(len(x))])
plt.plot(Ca[0,:],Ca[1,:], color='navy', linewidth=2)
plt.plot(Cb[0,:],Cb[1,:], color='navy', linewidth=2)
plt.plot(Cc[0,:],Cc[1,:], color='blue', linewidth=2)
RfM = np.array([[ 1, 0, 0], [ 0,-1, 0], [ 0, 0, 1]])
Ca2 = RfM @ Ca
Cb2 = RfM @ Cb
Cc2 = RfM @ Cc
plt.plot(Ca2[0,:],Ca2[1,:], color='navy', linewidth=2)
plt.plot(Cb2[0,:],Cb2[1,:], color='navy', linewidth=2)
plt.plot(Cc2[0,:],Cc2[1,:], color='blue', linewidth=2)
plt.axis('equal')
```
